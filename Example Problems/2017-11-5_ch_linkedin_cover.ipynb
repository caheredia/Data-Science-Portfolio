```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline 
```

## This notebook will output a pdf graphic with the appropitate dimensions for a Linkedin cover photo. Current ratio is 4:1.
The graph will be a sigmoid function with the axes Execution and Time. The plot will have annoted text with 'ideation', 'development', and 'delivery'

```python
def sigmoid(X):
    return 1/(1+np.exp(-X))
```

```python
x = np.arange(-10,10,.01)
y = sigmoid(x)
```

```python
from matplotlib import rcParams

#set plot attributes
fig_width = 8  # width in inches
fig_height = 2  # height in inches
fig_size =  [fig_width,fig_height]
params = {'backend': 'Agg',
          'axes.labelsize': 8,
          'axes.titlesize': 8,
          'font.size': 10,
          'xtick.labelsize': 8,
          'ytick.labelsize': 8,
          'figure.figsize': fig_size,
          'savefig.dpi' : 300,
          'font.family': 'sans-serif',
          'axes.linewidth' : 0.5,
          'font.size' : 8
          }
rcParams.update(params)

#adjust subplots position
fig = plt.figure()

ax1 = plt.axes((0, 0, 1, 1))
plt.plot(x, y, 'orange', lw=1)


plt.annotate(
    'Ideation', fontsize = 12, 
    xy=(-9,0), arrowprops=dict(arrowstyle='-'), xytext=(-9.75, .2))
plt.annotate(
    'Exploration', fontsize = 12, 
    xy=(0,.5), arrowprops=dict(arrowstyle='-'), xytext=(-3, .475))
plt.annotate(
    'Delivery', fontsize = 12, 
    xy=(9,1), arrowprops=dict(arrowstyle='-'), xytext=(8.25,.8))


#set axis limits
ax1.set_xlim((x.min(), x.max()))
ax1.set_ylim((y.min()-.1, y.max()+.1))

#hide right and top axes
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.spines['bottom'].set_position(('outward', 1))
ax1.spines['left'].set_position(('outward', 1))
ax1.yaxis.set_visible(False)
ax1.xaxis.set_visible(False)

ax1.tick_params(axis='both', which='both', bottom='off', left='off', labelbottom='off', labelleft='off') 

#set labels
plt.xlabel('Time')
plt.ylabel('Execution')

#export to svg
plt.savefig('sigmoid_fit.png', transparent=False)
```
